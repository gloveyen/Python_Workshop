### Pytrends - Google Trends Datasets
google關鍵字搜尋熱度資料

```python
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
```

```python
pytrend.build_payload(kw_list=['data science','machine learning','deep learning'], cat=0, timeframe='today 12-m', geo='TW', gprop='')
df = pytrend.interest_over_time()
```

```python
df.plot(figsize=(10,6))
plt.show()
```
